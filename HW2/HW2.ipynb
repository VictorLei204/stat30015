### 1.(1) Prove the Gauss-Markov Theorem

令 $\tilde \beta = Cy$ 为任意y的线性估计，其中 C 是一个 $p \times N$ 矩阵，再设 $C = (X^TX)^{-1}X^T + D$，其中 D 是一个 $p \times N$ 矩阵。
$$
\begin{aligned}
E[\tilde{\beta}] &= E[Cy] \\
&= E[((X^TX)^{-1}X^T + D)(X\beta+\epsilon)] \\
&= ((X^TX)^{-1}X^T+D)X\beta + ((X^TX)^{-1}X^T + D)E[\epsilon]  \\
&= \beta + DX\beta\\
\end{aligned}
$$
要使 $\tilde \beta$ 为无偏估计，则 $DX = 0_{p\times p}$
$$
\begin{aligned}
Var(\tilde{\beta}) &= Var(Cy) \\
&= CVar(y)C^T \\
&= \sigma^2 CC^T \\
&= \sigma^2 ((X^TX)^{-1}X^T + D)((X^TX)^{-1}X^T + D)^T \\
&= \sigma^2 (X^TX)^{-1} + \sigma^2 (X^TX)^{-1}(DX)^T + \sigma^2 DX(X^TX)^{-1} + \sigma^2 DD^T \\
&= \sigma^2 (X^TX)^{-1} + \sigma^2 DD^T \\
&= Var(\hat{\beta}) + \sigma^2 DD^T \\
&\ge Var(\hat{\beta})
\end{aligned}
$$
其中 $\hat{\beta} = (X^TX)^{-1}X^Ty$ 是OLS估计量，$DD^T$ 是半正定矩阵，则当且仅当 $D=0_{p\times N}$ 时，$Var(\tilde{\beta})=Var(\hat{\beta})$ ，此时 $\tilde{\beta}=\hat{\beta}$ ，$\hat{\beta}$ 为 $\beta$ 的最优线性无偏估计。


### 1.(2.1)



$$
\begin{aligned}
y &= X \beta + \epsilon \\
y - X \beta &= \epsilon \sim N(0,\sigma^2 I_N) \\
\end{aligned}\\
\text{}\\
\begin{aligned}
L(\beta) &= \frac{1}{\sqrt{(2\pi)^Ndet(\sigma^2 I_N)}}\exp(-\frac{(y - X \beta)^T(\sigma^2 I_N)^{-1}(y - X \beta)}{2}) \\
&= \frac{1}{\sqrt{(2\pi\sigma^2)^N}}\exp(-\frac{(y - X \beta)^T(y - X \beta)}{2\sigma^2}) \\
\end{aligned}\\
\ln L(\beta) = -\frac{N}{2}\ln(2\pi\sigma^2) - \frac{(y - X \beta)^T(y - X \beta)}{2\sigma^2} \\
\text{}\\
\frac{\partial \ln L(\beta)}{\partial \beta} = -(\frac{\partial (y - X \beta)}{\partial \beta})^T\frac{(y - X \beta)}{\sigma^2} = \frac{1}{\sigma^2}X^T(y - X \beta)
\text{}\\
\frac{1}{\sigma^2}X^T(y - X \hat{\beta}_{MLE}) = 0 \\
\text{}\\
\hat{\beta}_{MLE} = (X^TX)^{-1}X^Ty = \hat{\beta}_{OLS}
$$

### 1.(2.2)

$$
\begin{aligned}
E(\hat{\sigma}^2) &= E[\frac{(y-X\hat{\beta})^T(y-X\hat{\beta})}{N-p}] \\
&= \frac{1}{N-p}E[(y-X(X^TX)^{-1}X^Ty)^T(y-X(X^TX)^{-1}X^Ty)] \\
&= \frac{1}{N-p}E[y^Ty - y^TX(X^TX)^{-1}X^Ty] \\
&= \frac{1}{N-p}E[(X\beta + \epsilon)^T(I_N - X(X^TX)^{-1}X^T)(X\beta + \epsilon)] \\
&= \frac{1}{N-p}E[\epsilon^T\epsilon - \epsilon^TX(X^TX)^{-1}X^T\epsilon + \epsilon^T(I_N - X(X^TX)^{-1}X^T)X\beta + \beta^TX^T(I_N - X(X^TX)^{-1}X^T)\epsilon] \\
&= \frac{1}{N-p}E[\epsilon^T(I-X(X^TX)^{-1}X^T)\epsilon] \quad \text{其中}(I-X(X^TX)^{-1}X^T)\text{为对称矩阵，利用二次型期望公式得}\\
&= \frac{1}{N-p}(tr[(I-X(X^TX)^{-1}X^T)\sigma^2 I_N] + 0^T(I-X(X^TX)^{-1}X^T)0) \\
&= \frac{\sigma^2}{N-p}tr(I-X(X^TX)^{-1}X^T) \\
&= \frac{\sigma^2}{N-p}(N - tr(X^TX(X^TX)^{-1})) \\
&= \frac{\sigma^2}{N-p}(N - p) \\
&= \sigma^2
\end{aligned}
$$

### 1.(2.3)

#### (1)

$$
\begin{aligned}
\hat{\beta} &= (X^TX)^{-1}X^Ty\\
&= (X^TX)^{-1}X^T(X\beta + \epsilon)\\
&= \beta + (X^TX)^{-1}X^T\epsilon
\end{aligned}
$$
$\hat{\beta}$是关于 $\epsilon$ 的线性变换，服从正态分布
$$
E(\hat{\beta}) = \beta + (X^TX)^{-1}X^TE(\epsilon) = \beta
$$
$$
\begin{aligned}
Var(\hat{\beta}) &= (X^TX)^{-1}X^TVar(\epsilon)X(X^TX)^{-1} \\
&= (X^TX)^{-1}X^T\sigma^2 I_N X(X^TX)^{-1} \\
&= (X^TX)^{-1} \sigma^2
\end{aligned}
$$
$$\hat{\beta} \sim N (\beta, (X^TX)^{-1}\sigma^2)$$

#### (2)

$$
\begin{aligned}
(N-p)\hat{\sigma}^2 &= (y-X\hat{\beta})^T(y-X\hat{\beta}) \\
&= \epsilon^T(I_N - X(X^TX)^{-1}X^T)\epsilon
\end{aligned}
$$

$$(I_N - X(X^TX)^{-1}X^T)^2 = I_N - X(X^TX)^{-1}X^T$$
$$(I_N - X(X^TX)^{-1}X^T)^T = I_N - X(X^TX)^{-1}X^T$$
$$
\begin{aligned}
rank(I_N - X(X^TX)^{-1}X^T) &= tr(I_N - X(X^TX)^{-1}X^T)\\
&= N - p
\end{aligned}
$$
$I_N - X(X^TX)^{-1}X^T$ 是秩和迹为 N-p 的对称幂等矩阵，则存在正交矩阵 P 使得
$$P^T(I_N - X(X^TX)^{-1}X^T)P = \begin{pmatrix}
I_{N-p} & 0 \\
0 & 0
\end{pmatrix}$$
令 $Z = P^T\epsilon$，则 $E(Z) = P^TE(\epsilon) = 0$，$Var(Z) = P^T\sigma^2I_NP = \sigma^2 I_N$，$Z \sim N(0, \sigma^2 I_N)$
$$
\begin{aligned}
(N-p)\hat{\sigma}^2 &= \epsilon^T(I_N - X(X^TX)^{-1}X^T)\epsilon \\
&= \epsilon^TPP^T(I_N - X(X^TX)^{-1}X^T)PP^T\epsilon \\
&= Z^T\begin{pmatrix}
I_{N-p} & 0 \\
0 & 0
\end{pmatrix}Z \\
&= \sum_{i=1}^{N-p}Z_i^2
\end{aligned}
$$
则 $(N-p)\hat{\sigma}^2 = \sigma^2\sum_{i=1}^{N-p}(\frac{Z_i}{\sigma})^2 \sim \sigma^2\chi^2_{N-p}$
