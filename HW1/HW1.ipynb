## 1

$$
(\exp(sx))' = s \cdot \exp(sx)
$$
$$
(\exp(sx))'' = s^2 \cdot \exp(sx) > 0
$$
$\exp(sx)$为关于x的凸函数，得
$$
\forall x \in [a,b], \quad \exp(sx) \leq \frac{b-x}{b-a}\exp(sa) + \frac{x-a}{b-a}\exp(sb)
$$
$$
E[\exp(sX)] \leq E\left[\frac{b-X}{b-a}\exp(sa) + \frac{X-a}{b-a}\exp(sb)\right]
$$
$$
E[\exp(sX)] \leq \frac{b-E[X]}{b-a}\exp(sa) + \frac{E[X]-a}{b-a}\exp(sb)
$$
$$
E[\exp(sX)] \leq \frac{b}{b-a}\exp(sa) - \frac{a}{b-a}\exp(sb)
$$
令$t = s(b-a)$，
$$
E[\exp(sX)] \leq \exp (\frac{ta}{b-a} + \ln(\frac{b}{b-a} - \frac{a}{b-a}\exp(t)))
$$
令$f(t) = \frac{ta}{b-a} + \ln(\frac{b}{b-a} - \frac{a}{b-a}\exp(t))$，
$$
f'(t) = \frac{a}{b-a} + \frac{-\frac{a}{b-a}\exp(t)}{\frac{b}{b-a} - \frac{a}{b-a}\exp(t)} = \frac{a}{b-a} - \frac{a \exp(t)}{b - a \exp(t)}
$$
$$
f''(t) = \frac{-ab \exp(t)}{(b - a \exp(t))^2} \leq \frac{1}{4}
$$
$$f(0)=0, \quad f'(0)=0$$
$$
f(t) = f(0) + f'(0)t + \frac{f''(\theta t)}{2}t^2 \leq \frac{1}{8}t^2 \quad \theta \in [0,1]
$$
$$
E[\exp(sX)] \leq \exp(f(t)) \leq \exp(t^2/8) = \exp[s^2(b-a)^2/8]
$$

## 2

### a

$$y_i = \sum_{k=1}^{n} A_{ik} x_k$$
$$\frac{\partial y_i}{\partial x_j} = \frac{\partial}{\partial x_j} \left( \sum_{k=1}^{n} A_{ik} x_k \right)$$
$$\frac{\partial y_i}{\partial x_j} = \sum_{k=1}^{n} A_{ik} \frac{\partial x_k}{\partial x_j}$$
$$\frac{\partial y_i}{\partial x_j} = A_{ij}$$
$$(\frac{\partial \mathbf{y}}{\partial \mathbf{x}^\top})_{ij} = A_{ij}$$
$$\frac{\partial \mathbf{y}}{\partial \mathbf{x}^\top} = \mathbf{A}$$

### b

$$\alpha = \sum_{i=1}^{m} y_i (\mathbf{A}\mathbf{x})_i = \sum_{i=1}^{m} y_i \left( \sum_{j=1}^{n} A_{ij} x_j \right)$$

$$\frac{\partial \alpha}{\partial x_k} = \frac{\partial}{\partial x_k} \sum_{i=1}^{m} \sum_{j=1}^{n} y_i A_{ij} x_j = \sum_{i=1}^{m} \sum_{j=1}^{n} \frac{\partial}{\partial x_k} (y_i A_{ij} x_j)$$   $$\frac{\partial \alpha}{\partial x_k} = \sum_{i=1}^{m} \sum_{j=1}^{n} \left[ \left(\frac{\partial y_i}{\partial x_k}\right) (A_{ij} x_j) + y_i \left(A_{ij} \frac{\partial x_j}{\partial x_k}\right) \right]$$

$$\sum_{i=1}^{m} \sum_{j=1}^{n} \left(\frac{\partial y_i}{\partial x_k}\right) A_{ij} x_j =(\mathbf{x}^\top\mathbf{A}^\top \frac{\partial \mathbf{y}}{\partial \mathbf{x}^\top})_k$$
$$\sum_{i=1}^{m} \sum_{j=1}^{n} y_i A_{ij} \frac{\partial x_j}{\partial x_k}=\sum_{i=1}^{m} y_i A_{ik}=(\mathbf{y}^\top\mathbf{A})_k$$

$$\frac{\partial \alpha}{\partial \mathbf{x}^\top} = \mathbf{x}^\top\mathbf{A}^\top \left(\frac{\partial \mathbf{y}}{\partial \mathbf{x}^\top}\right) + \mathbf{y}^\top\mathbf{A}$$

### c

由b结论得，令$y=x$
$$\frac{\partial \alpha}{\partial \mathbf{x}^\top} = \mathbf{x}^\top\mathbf{A}^\top \left(\frac{\partial \mathbf{x}}{\partial \mathbf{x}^\top}\right) + \mathbf{x}^\top\mathbf{A}$$

$$(\frac{\partial \mathbf{x}}{\partial \mathbf{x}^\top})_{ij} = \frac{\partial x_i}{\partial x_j} =\begin{cases} 
1 & i=j \\ 
0 & otherwise 
\end{cases}$$
$$\frac{\partial \mathbf{x}}{\partial \mathbf{x}^\top} = \mathbf{I}$$

$$\frac{\partial \alpha}{\partial \mathbf{x}^\top} = \mathbf{x}^\top\mathbf{A}^\top\mathbf{I} + \mathbf{x}^\top\mathbf{A} = \mathbf{x}^\top(\mathbf{A}^\top + \mathbf{A})$$

### d

$$\frac{\partial \alpha}{\partial \mathbf{z}^\top} = \frac{\partial \alpha}{\partial \mathbf{y}^\top} \frac{\partial \mathbf{y}}{\partial \mathbf{z}^\top} + \frac{\partial \alpha}{\partial \mathbf{x}^\top} \frac{\partial \mathbf{x}}{\partial \mathbf{z}^\top}$$

由$\alpha = \mathbf{y}^\top\mathbf{A}\mathbf{x}$和a结论得，$\frac{\partial \alpha}{\partial \mathbf{x}^\top} = \mathbf{y}^\top\mathbf{A}$，
$\frac{\partial \alpha}{\partial \mathbf{y}^\top} = (\mathbf{A}\mathbf{x})^\top = \mathbf{x}^\top\mathbf{A}^\top$

$$\frac{\partial \alpha}{\partial \mathbf{z}^\top} = \mathbf{x}^\top\mathbf{A}^\top \left(\frac{\partial \mathbf{y}}{\partial \mathbf{z}^\top}\right) + \mathbf{y}^\top\mathbf{A} \left(\frac{\partial \mathbf{x}}{\partial \mathbf{z}^\top}\right)$$

### e

$$\mathbf{A}\mathbf{A}^{-1} = \mathbf{I}$$
$$\frac{\partial}{\partial \alpha}(\mathbf{A}\mathbf{A}^{-1}) = \frac{\partial \mathbf{I}}{\partial \alpha} = \mathbf{0}$$
$$\left(\frac{\partial \mathbf{A}}{\partial \alpha}\right) \mathbf{A}^{-1} + \mathbf{A} \left(\frac{\partial \mathbf{A}^{-1}}{\partial \alpha}\right) = \mathbf{0}$$
$$\mathbf{A} \left(\frac{\partial \mathbf{A}^{-1}}{\partial \alpha}\right) = -\left(\frac{\partial \mathbf{A}}{\partial \alpha}\right) \mathbf{A}^{-1}$$
$$\mathbf{A}^{-1}\mathbf{A} \left(\frac{\partial \mathbf{A}^{-1}}{\partial \alpha}\right) = -\mathbf{A}^{-1}\left(\frac{\partial \mathbf{A}}{\partial \alpha}\right) \mathbf{A}^{-1}$$
$$\frac{\partial \mathbf{A}^{-1}}{\partial \alpha} = -\mathbf{A}^{-1}\frac{\partial \mathbf{A}}{\partial \alpha}\mathbf{A}^{-1}$$
