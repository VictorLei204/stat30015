### 1.

$$
\begin{aligned}
L &= \prod_{i=1}^{N}P(X=x_i, Y=y_i)\\
    &= \prod_{i=1}^{N}P(X=x_i|Y=y_i)P(Y=y_i)\\
    &= \prod_{i=1}^{N}(\prod_{j=1}^{n}P(X^{(j)}=x_{i}^{(j)}|Y=y_i))P(Y=y_i)\\
\log L &= \sum_{i=1}^{N}\log P(Y=y_i) + \sum_{i=1}^{N}\sum_{j=1}^{n}\log P(X^{(j)}=x_{i}^{(j)}|Y=y_i)\\
    &= \sum_{i=1}^{N}\sum_{k=1}^{K}I(y_i=c_k)\log P(Y=c_k) + \sum_{i=1}^{N}\sum_{j=1}^{n}\sum_{l=1}^{S_j}I(x_i^{(j)}=a_{jl})\log P(X^{(j)}=a_{jl}|Y=y_i)\\
    &= \sum_{i=1}^{N}\sum_{k=1}^{K}I(y_i=c_k)\log P(Y=c_k) + \sum_{i=1}^{N}\sum_{j=1}^{n}\sum_{l=1}^{S_j}\sum_{k=1}^KI(x_i^{(j)}=a_{jl},y_i = c_k)\log P(X^{(j)}=a_{jl}|Y=c_k)\\
\end{aligned}\\
\begin{aligned}
\frac{\partial \log L - \lambda(\sum_{k=1}^{K}P(Y=c_k)-1)}{\partial P(Y=c_k)} &= \sum_{i=1}^{N}\frac{I(y_i=c_k)}{P(Y=c_k)} - \lambda = 0 \quad P(Y=c_k)=\frac{\sum_{i=1}^{N}I(y_i=c_k)}{\lambda}\\
\frac{\partial \log L - \lambda(\sum_{k=1}^{K}P(Y=c_k)-1)}{\partial \lambda} &= \sum_{k=1}^{K}P(Y=c_k)-1 = 0\\
\text{解得}\hat{P}(Y=c_k) &= \frac{\sum_{i=1}^{N}I(y_i=c_k)}{N}\quad \lambda = N\\
\end{aligned}\\
\begin{aligned}
\frac{\partial \log L - \sum_{j=1}^n\sum_{k=1}^{K}\lambda_{jk}(\sum_{l=1}^{S_j}P(X^{(j)}=a_{jl}|Y=c_k)-1)}{\partial P(X^{(j)}=a_{jl}|Y=c_k)} &= \sum_{i=1}^{N}\frac{I(x_i^{(j)}=a_{jl},y_i = c_k)}{P(X^{(j)}=a_{jl}|Y=c_k)} - \lambda_{jk} = 0 \quad P(X^{(j)}=a_{jl}|Y=c_k)=\frac{\sum_{i=1}^{N}I(x_i^{(j)}=a_{jl},y_i = c_k)}{\lambda_{jk}}\\
\sum_{l=1}^{S_j}P(X^{(j)}=a_{jl}|Y=c_k)-1 &= 0\\
\text{解得}\hat{P}(X^{(j)}=a_{jl}|Y=c_k) &= \frac{\sum_{i=1}^{N}I(x_i^{(j)}=a_{jl},y_i = c_k)}{\sum_{i=1}^{N}I(y_i=c_k)}\quad \lambda_{jk} = \sum_{i=1}^{N}I(y_i=c_k)\\
\end{aligned}
$$

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams["font.family"] = "Noto Sans SC"

train_set: pd.DataFrame = pd.read_csv("./train_set.csv")
test_set: pd.DataFrame = pd.read_csv("./test_set.csv")
```

```python
print(
    train_set["单位名称"]
    .value_counts()
    .to_frame()
    .sort_values("count", ascending=False)
)
plot = (
    train_set["单位名称"]
    .value_counts()
    .to_frame()
    .sort_values("count", ascending=False)
    .plot(
        kind="bar",
        title="各单位名称出现频次柱状图",
        rot=0,
        figsize=(8, 6),
        legend=False,
    )
)
```

根据统计结果，可以清晰地看到市民投诉在不同政府单位间的分布存在明显分层：

第一梯队： “市水务集团”（557件）。

第二梯队： “市供热公司”（332件）、“市运输管理局”（330件）和“市燃气集团”（285件）的投诉量处于中游，且前两者的数量非常接近。

第三梯队： “市公交集团”（207件）的投诉量明显低于第二梯队。

第四梯队： “市房地集团”（96件）和“市供电公司”（93件）的投诉量最少，均不足100件。

“市供热公司”的投诉量高居第二，和数据收集于冬季相符合。

```python
print(train_set.sum(axis=1, numeric_only=True).describe())
plot = train_set.sum(axis=1, numeric_only=True).hist()
```

用词数呈现显著的右偏分布和厚尾分布：

绝大多数投诉的用词数都比较少，少数投诉包含了非常多的词汇。

```python
train_set["word_count"] = train_set.sum(axis=1, numeric_only=True)
plot = train_set.boxplot(column="word_count", by="单位名称")
train_set.drop(columns=["word_count"], inplace=True)
```

各部门的用词数中位数基本一致，各部门大部分市民的投诉篇幅都差不多。

除房地产部门仅有一离群点外，其他部门都存在大量的离群点,虽然大部分投诉比较简短，但所有部门都会收到用词数极多的投诉。

和第一问相比较，可以看出投诉量大的部门，用词数极多的投诉也多。

```python
# 将词频转为是否使用该词
train_set[train_set.columns[1:]] = (train_set[train_set.columns[1:]] > 0).astype("int")
test_set[test_set.columns[1:]] = (test_set[test_set.columns[1:]] > 0).astype("int")
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

clf = BernoulliNB()
clf.fit(
    X=train_set.drop(columns=["单位名称"]),
    y=train_set["单位名称"],
)
y_pred = clf.predict(X=test_set.drop(columns=["单位名称"]))
cm = confusion_matrix(test_set["单位名称"], y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
fig, ax = plt.subplots(figsize=(10, 10))
disp.plot(ax=ax)
plt.show()
```

模型总体有效，但对各部门具体效果有很大差异：

1. 水务部门、公交部门、运输管理局、燃气集团、供热公司表现良好，对角线上的值高于其他值，说明模型能较好地区分这些部门的投诉。
2. 而对于房地集团和供电公司则完全失败，正确预测数为0，说明模型无法区分这些部门的投诉，可能是因为这些部门的投诉量过少，导致训练数据不足。
